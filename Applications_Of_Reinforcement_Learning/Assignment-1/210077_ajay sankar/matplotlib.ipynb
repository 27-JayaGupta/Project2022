```python
from matplotlib import pyplot as plt
import numpy as np
```

### pie chart


```python
y = np.array([29, 22, 14, 6])
mylabels = ["packaging", "food containers", "fishing and beach items", "glass bottles"]

plt.pie(y, labels = mylabels)
plt.title("top categories of polluters in oceans")
plt.show()
```

### line graph


```python
x1 = np.array(['day 1', 'day 2', 'day 3', 'day 4'])
y1 = np.array([7, 6, 9, 8])
y2 = np.array([8, 2, 6, 9])
y3 = np.array([2, 5, 3, 7])

plt.plot(x1,y1, marker='x')
plt.plot(x1,y2, marker='s')
plt.plot(x1,y3, marker='>')
plt.title("cups of coffee per employee")
plt.show()
```
